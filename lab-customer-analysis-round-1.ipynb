```python
#Renaming file1 & 2 columns to match file 3
file1 = file1.rename(columns={ 'ST':'State','GENDER':'Gender'})
file2 = file2.rename(columns={ 'ST':'State','GENDER':'Gender'})
```

```python
#Rearranging columns in file2 to match file1
file2 = file2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#Rearranging columns in file3 to match file1 & 2
file3 = file3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
#Concatenate the three dataframes
customer_analysis = pd.concat([file1, file2, file3], axis=0)
customer_analysis.shape
```

```python
customer_analysis.dtypes
```

```python
#Income, Monthly Premium Auto and Tota Claim Amount are numerical
#Customer, State, Gender, Education, Customer Lifetime Value*, Number of Open Complaints*, Policy Type, Vehicle Class are categorical
#Customer Lifetime Value*, Number of Open Complaints* should be numerical (float64 and int64)
```

```python
#Delete the column education and the number of open complaints from the dataframe.
customer_analysis = customer_analysis.drop(['Education', 'Number of Open Complaints'], axis = 1)
customer_analysis.head()
```

```python
customer_analysis.shape
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 (?) and change dtype to numerical type.
def clean(x):
    x = str(x)
    if x.endswith("%"):
        x = x.replace("%","")
    return float(x)

customer_analysis['Customer Lifetime Value'] = list(map(clean, customer_analysis['Customer Lifetime Value']))
customer_analysis['Customer Lifetime Value'].unique()
customer_analysis.head(20)
```

```python
customer_analysis.dtypes
```

```python
# Check for duplicate rows in the data and remove if any.
customer_analysis = customer_analysis.copy()
customer_analysis = customer_analysis.drop_duplicates()  # nb keep argument
print(customer_analysis.shape)
```
